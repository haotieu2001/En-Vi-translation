```python
import gradio as gr 
from transformers import pipeline 
trans = pipeline("translation", model="haotieu/en-vi-mt-model")
def translation_gradio(input):
  result = trans(input)
  return result[0]['translation_text']

translation_interface = gr.Interface(fn = translation_gradio, inputs="text", outputs = "text",title="english-vietnamese translation", examples=["Hi my name is Hao"])
```

```python
translation_interface.launch(debug=True)
```
